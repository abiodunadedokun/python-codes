```python
import fitz

# Open the PDF file
pdf_file = "ex6.pdf"
with fitz.open(pdf_file) as doc:
    # Extract the text from the first page
    page = doc.load_page(0)
    text = page.get_text()

    # Preprocess the text
    text = text.replace('\n', ' ')
    text = text.replace('\t', ' ')
    text = ' '.join(text.split())

    # Get the bounding boxes of the characters
    blocks = page.get_text_blocks()
    boxes = []
    for b in blocks:
        x0, y0, x1, y1 = b[:4]
        for char in b[4]:
            boxes.append((char, (x0, y0, x1, y1)))

    # Print the text and bounding boxes
    #print(text)
    print(boxes)
```
