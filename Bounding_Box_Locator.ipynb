```python
import pdfplumber

def extract_text_and_bboxes_from_pdf(pdf_path):
    text_with_bboxes = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_content = page.extract_text()
            text_with_bboxes.append({"text": page_content, "bboxes": []})

            for obj in page.chars:
                bbox = [obj["x0"], obj["top"], obj["x1"], obj["bottom"]]
                text_with_bboxes[-1]["bboxes"].append(bbox)

    return text_with_bboxes

if __name__ == "__main__":
    pdf_file_path = r"example.pdf"
    extracted_data = extract_text_and_bboxes_from_pdf(pdf_file_path)
    for page_data in extracted_data:
        print("Page Text:")
        print(page_data["text"])
        print("Bounding Boxes:")
        for bbox in page_data["bboxes"]:
            print(f"x1: {bbox[0]}, y1: {bbox[1]}, x2: {bbox[2]}, y2: {bbox[3]}")
        print("\n")
```

```python
import pdfplumber

def extract_text_and_bboxes_from_pdf(pdf_path):
    text_with_bboxes = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_content = page.extract_text()
            bboxes_data = []

            for obj in page.chars:
                bbox = [obj["x0"], obj["top"], obj["x1"], obj["bottom"]]
                char_text = obj["text"]
                bboxes_data.append((bbox, char_text))

            text_with_bboxes.append({"text": page_content, "bboxes_data": bboxes_data})

    return text_with_bboxes

if __name__ == "__main__":
    pdf_file_path = r"example.pdf"
    extracted_data = extract_text_and_bboxes_from_pdf(pdf_file_path)
    for page_data in extracted_data:
        print("Page Text:")
        print(page_data["text"])
        print("Bounding Boxes with Characters:")
        for bbox, char_text in page_data["bboxes_data"]:
            print(f"Character: '{char_text}', Bounding Box: x1: {bbox[0]}, y1: {bbox[1]}, x2: {bbox[2]}, y2: {bbox[3]}")
        print("\n")
```

```python
import pdfplumber

def extract_text_and_bboxes_from_pdf(pdf_path):
    text_with_bboxes = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_content = page.extract_text()
            bboxes_data = []

            for obj in page.chars:
                bbox = [obj["x0"], obj["top"], obj["x1"], obj["bottom"]]
                char_text = obj["text"]
                unicode_code_point = ord(char_text)
                bboxes_data.append((bbox, char_text, unicode_code_point))

            text_with_bboxes.append({"text": page_content, "bboxes_data": bboxes_data})

    return text_with_bboxes

if __name__ == "__main__":
    pdf_file_path = r"c:/Users/Acer/Desktop/Emden76-1.pdf"
    extracted_data = extract_text_and_bboxes_from_pdf(pdf_file_path)
    for page_data in extracted_data:
        print("Page Text:")
        print(page_data["text"])
        print("Bounding Boxes with Characters and Unicode Code Points:")
        for bbox, char_text, unicode_code_point in page_data["bboxes_data"]:
            print(f"Character: '{char_text}', Unicode Code Point: U+{unicode_code_point:04X}, Bounding Box: x1: {bbox[0]}, y1: {bbox[1]}, x2: {bbox[2]}, y2: {bbox[3]}")
        print("\n")
```
